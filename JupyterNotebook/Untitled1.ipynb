```python
from langchain_community.tools.sql_database.tool import QuerySQLDataBaseTool
from langchain_community.utilities import SQLDatabase
from langchain_community.agent_toolkits import create_sql_agent
from langchain_google_genai import ChatGoogleGenerativeAI
import os


# --- API KEY ---
os.environ["GOOGLE_API_KEY"] = "" # add your api key

# --- DATABASE ---
db = SQLDatabase.from_uri(
    "mysql+pymysql://root:Nurdos9900!@localhost/bank_marketing"
)
tool = QuerySQLDataBaseTool(db=db)
# --- LLM ---
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature=0
)

# --- AGENT ---
agent_executor = create_sql_agent(
    llm=llm,
    db=db,
    verbose=True,
    handle_parsing_errors=True
)

# --- RUN ---
response = agent_executor.run("Сколько строк во всех таблицах?")
print(response)
```

```python
response = agent_executor.run("Identify days/months/times when conversion is highest to reduce wasted calls.")
print(response)
```
